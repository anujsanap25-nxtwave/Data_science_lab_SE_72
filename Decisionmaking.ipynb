```python
score = 85 # You can change this score to test different outcomes

if score >= 90:
    grade = 'A'
elif score >= 75:
    grade = 'B'
elif score >= 60:
    grade = 'C'
else:
    grade = 'F'

print(f"A student with a score of {score} receives a grade of: {grade}")
```
